```python
data = tpi
data = time_format(data,'DocRef1EndDate','fmt_DocRef1EndDate')
data = time_year(data,'fmt_DocRef1EndDate','year(DocRef1EndDate)')
data = filter_col(data,'Ref1Category','DocRef1Name')

TPI_Ref_2 = group(data,'ResearchCycleId','FTSE_ID','ResearchId','fmt_DocRef1EndDate','year(DocRef1EndDate)')

TPI_Ref_2=TPI_Ref_2.rename(columns={"fmt_DocRef1EndDate": "Date","year(DocRef1EndDate)":"Year(Date)"})
TPI_Ref_2['Method']= 'DocRef1EndDate'

present = pd.concat([TPI_Ref_2['ResearchId']])
remaining_df = tpi[~tpi['ResearchId'].isin(present)]
```

```python
data = remaining_df
data = time_format(data,'DocRef2EndDate','fmt_DocRef2EndDate')
data = time_year(data,'fmt_DocRef2EndDate','year(DocRef2EndDate)')
data = filter_col(data,'Ref2Category','DocRef2Name')

TPI_Ref_3 = group(data,'ResearchCycleId','FTSE_ID','ResearchId','fmt_DocRef2EndDate','year(DocRef2EndDate)')

TPI_Ref_3=TPI_Ref_3.rename(columns={"fmt_DocRef2EndDate": "Date","year(DocRef2EndDate)":"Year(Date)"})
TPI_Ref_3['Method']= 'DocRef2EndDate'

present = pd.concat([TPI_Ref_2['ResearchId'],TPI_Ref_3['ResearchId']])
remaining_df = tpi[~tpi['ResearchId'].isin(present)]
```
